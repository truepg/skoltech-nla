_1._ (10 pts) Consider the following function

$$ F(U, V) = \frac{1}{2}\|X - UV\|_F^2, $$

_2._ (2 pts) Derive analytical expression for the gradient of the function $f$:

$$ R(x) = \frac{(Ax, x)}{(x, x)}, $$

where $A$ is a symmetric real matrix. Why the gradient of this function is important in NLA you will know in the lectures later.

### 1.1
$F(U, V) = \frac{1}{2}\|X - UV\|_F^2\\$
$W = UV = {w_{ij}=\sum_{m=1}^{k}{u_{im}v_{mj}}}$

$F(U, V) = \frac{1}{2}\|X - UV\|_F^2 = \sum_{i=1}^n \sum_{j=1}^n (x_{ij} - \sum_{m=1}^{k}{u_{im}v_{mj}})^2$

$\frac{\partial F}{\partial u_{im}} = \sum_{j=1}^n(x_{ij} - \sum_{m=1}^{k}{u_{im}v_{mj}})(-v_{mj}) = - \sum_{j=1}^n(x_{ij} - \sum_{m=1}^{k}{u_{im}v_{mj}})v_{mj} => \frac{\partial F}{\partial U} = -(X-UV)V^T$

### 1.2

$F(U, V) = \frac{1}{2}\|X - UV\|_F^2 = \sum_{i=1}^n \sum_{j=1}^n (x_{ij} - \sum_{m=1}^{k}{u_{im}v_{mj}})^2$

$\frac{\partial F}{\partial v_{mj}} = \sum_{i=1}^n(x_{ij} - \sum_{m=1}^{k}{u_{im}v_{mj}})(-u_{im}) = - \sum_{i=1}^n u_{im}(x_{ij} - \sum_{m=1}^{k}{u_{im}v_{mj}}) => \frac{\partial F}{\partial V} = -U^T(X-UV)$

### 1.3

First matmul: $U_{n \times k}V_{k \times n} => O(k \cdot n^2)$ <br>
Second matmul: $(X-UV)_{n \times n}V_{n \times k}^T => O(k \cdot n^2)$ <br>
Gradient computational complexity = first matmul + second matmul = $O(k \cdot n^2) + O(k \cdot n^2)$ = $O(k \cdot n^2)$

````python
# Your solution is here

import jax
import jax.numpy as jnp
from jax.numpy.linalg import norm
from timeit import timeit
from tqdm.notebook import tqdm
````

````python
@jax.jit
def F(X, U, V):
    return 0.5 * jnp.linalg.norm(X - U @ V) ** 2

def manual_F_grad_U(X, U, V):
    return -(X - U @ V) @ V.T

def manual_F_grad_V(X, U, V):
    return -U.T @ (X - U @ V)
````

````python
F_grad_U = jax.jit(jax.grad(F, argnums=1, has_aux=False))
F_grad_V = jax.jit(jax.grad(F, argnums=2, has_aux=False))
````

````python
n = 10
X = jax.random.normal(jax.random.PRNGKey(0), (n, n))
U = jax.random.normal(jax.random.PRNGKey(1), (n, n))
V = jax.random.normal(jax.random.PRNGKey(2), (n, n))

norm(manual_F_grad_U(X, U, V) - F_grad_U(X, U, V)), norm(manual_F_grad_V(X, U, V) - F_grad_V(X, U, V))
````

````python
print("Compare speed")
print("Analytical gradient")
%timeit manual_F_grad_U(X, U, V)
print("Grad function")
%timeit F_grad_U(X, U, V).block_until_ready()
jit_F_grad_U = jax.jit(F_grad_U)
print("Jitted grad function")
%timeit jit_F_grad_U(X, U, V).block_until_ready()
````

### Gradient computation dependancy on dimension $n$

````python
k = 50
n_range = np.linspace(60, 1000, 10, dtype=int)

auto_grad_u_avg_time_lst_n = []
anal_grad_u_avg_time_lst_n = []
auto_grad_v_avg_time_lst_n = []
anal_grad_v_avg_time_lst_n = []

for n in tqdm(n_range):
    X = jax.random.normal(jax.random.PRNGKey(0), (n, n))
    U = jax.random.normal(jax.random.PRNGKey(1), (n, k))
    V = jax.random.normal(jax.random.PRNGKey(2), (k, n))
    
    anal_timeit_u = %timeit -oq manual_F_grad_U(X, U, V)
    anal_avg_time_u = anal_timeit_u.average
    
    auto_timeit_u = %timeit -oq F_grad_U(X, U, V).block_until_ready()
    auto_avg_time_u = auto_timeit_u.average
    
    anal_timeit_v = %timeit -oq manual_F_grad_V(X, U, V)
    anal_avg_time_v = anal_timeit_v.average
    
    auto_timeit_v = %timeit -oq F_grad_V(X, U, V).block_until_ready()
    auto_avg_time_v = auto_timeit_v.average
    
    auto_grad_u_avg_time_lst_n.append(auto_avg_time_u)
    anal_grad_u_avg_time_lst_n.append(anal_avg_time_u)
    auto_grad_v_avg_time_lst_n.append(auto_avg_time_v)
    anal_grad_v_avg_time_lst_n.append(anal_avg_time_v)
````

where $x$ is input data (in our case it will be images) and $w_i, \; i =1,\dots,3$ are parameters that are going to be trained. 

We will study the compression potential of neural network with simple architecture: alternating some numbers of linear and non-linear functions. 

The main task in this problem is to study how the compression of fully-connected layers affects the test accuracy.
Any fully-connected layer is represented as linear function $AX + B$, where $X$ is input matrix and $A, B$ are trainable matrices. Matrices $A$ in every layer are going to be compressed.

### Second step: neural network architecture

For simplicity and demonstration purposes of the neural network compression idea consider the architecture consisting of the only fully-connected layers and non-linear ReLU functions between them.
To demonstrate compression effect, consider the dimension of the inner layers equals to 1000.

Below you see implementation of such neural network in PyTorch.
More details about neural networks you will study in the *Deep learning* course in one of the upcoming term

````python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(3 * 32 * 32, 1000)
        self.fc2 = nn.Linear(1000, 1000)
        self.fc3 = nn.Linear(1000, 1000)
        self.fc4 = nn.Linear(1000, 1000)
        self.fc5 = nn.Linear(1000, 1000)
        self.fc6 = nn.Linear(1000, 10)
        self.ReLU = nn.ReLU()

    def forward(self, x):
        x = self.fc1(x.view(-1, 3 * 32*32))
        x = self.ReLU(x)
        x = self.fc2(x)
        x = self.ReLU(x)
        x = self.fc3(x)
        x = self.ReLU(x)
        x = self.fc4(x)
        x = self.ReLU(x)
        x = self.fc5(x)
        x = self.ReLU(x)
        x = self.fc6(x)
        return F.log_softmax(x, dim=1)
````

#### Implement functions for training and testing after every sweep over all dataset entries

````python
def train(model, train_loader, optimizer, epoch):
    model.train()
    for batch_idx, (data, target) in enumerate(train_loader):
        optimizer.zero_grad()
        output = model(data)
        loss = F.nll_loss(output, target)
        loss.backward()
        optimizer.step()
        if batch_idx % log_interval == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch, batch_idx * len(data), len(train_loader.dataset),
                100. * batch_idx / len(train_loader), loss.item()))
````

````python
def test(model, test_loader):
    model.eval()
    test_loss = 0
    correct = 0
    with torch.no_grad():
        for data, target in test_loader:
#             data, target = data.to(device), target.to(device)
            output = model(data)
            test_loss += F.nll_loss(output, target, reduction='sum').item() # sum up batch loss
            pred = output.argmax(dim=1, keepdim=True) # get the index of the max log-probability
            correct += pred.eq(target.view_as(pred)).sum().item()
    test_loss /= len(test_loader.dataset)

    print('\nTest set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
        test_loss, correct, len(test_loader.dataset),
        100. * correct / len(test_loader.dataset)))
````

### Set parameters for training and print intermediate loss values

````python
log_interval = 50
epochs = 7
````

### Third step: run training with the [Adam](https://arxiv.org/pdf/1412.6980.pdf%20%22%20entire%20document) optimization method

If your laptop is not very fast, you will wait some time till training is finished. 

````python
model = Net()
optimizer = optim.Adam(model.parameters(), lr=1e-3)

for epoch in range(1, epochs + 1):
    train(model,  train_loader, optimizer, epoch)
    test(model, test_loader)
````

````python
def layer_svd(layer, compression_rank):
    U, S, V = torch.svd(layer.weight.data)
    U_r, S_r, V_r = U[:,:compression_rank], S[:compression_rank], V[:,:compression_rank]
    
    layer_U_r = nn.Linear(compression_rank, 1000)
    layer_S_r = nn.Parameter(S_r, requires_grad = True)
    layer_V_r = nn.Linear(1000, compression_rank, bias=False)
    
    with torch.no_grad():
        layer_U_r.weight.copy_(U_r)
        layer_V_r.weight.copy_(V_r.T)
        
        layer_U_r.bias.copy_(layer.bias.data)
    
    return layer_U_r, layer_S_r, layer_V_r
````

````python
class CompressedNet(Net):
    def __init__(self,  model, compression_rank, layer_svd):
        super(CompressedNet, self).__init__()
        self.layer_svd = layer_svd
        self.compression_rank=compression_rank
        self.model = model
        self.ReLU = nn.ReLU()
        
        self.fc1 = model.fc1
        self.fc2_U, self.fc2_S, self.fc2_V = self.layer_svd(model.fc2, self.compression_rank)
        self.fc3_U, self.fc3_S, self.fc3_V = self.layer_svd(model.fc3, self.compression_rank)
        self.fc4_U, self.fc4_S, self.fc4_V = self.layer_svd(model.fc4, self.compression_rank)
        self.fc5_U, self.fc5_S, self.fc5_V = self.layer_svd(model.fc5, self.compression_rank)
        self.fc6 = model.fc6
       
        
    def forward(self, x):
        x = self.fc1(x.view(-1, 3 * 32*32))
        x = self.ReLU(x)
        x = self.fc2_U(torch.mul(self.fc2_S, self.fc2_V(x)))
        x = self.ReLU(x)
        x = self.fc3_U(torch.mul(self.fc3_S, self.fc3_V(x)))
        x = self.ReLU(x)
        x = self.fc4_U(torch.mul(self.fc4_S, self.fc4_V(x)))
        x = self.ReLU(x)
        x = self.fc5_U(torch.mul(self.fc5_S, self.fc5_V(x)))
        x = self.ReLU(x)
        x = self.fc6(x)
        x = self.ReLU(x)

        return F.log_softmax(x, dim=1)
````

### 3.3

````python
def test_accuracy(model, test_loader):
    model.eval()
    test_loss = 0
    correct = 0
    with torch.no_grad():
        for data, target in test_loader:
#             data, target = data.to(device), target.to(device)
            output = model(data)
            test_loss += F.nll_loss(output, target, reduction='sum').item() # sum up batch loss
            pred = output.argmax(dim=1, keepdim=True) # get the index of the max log-probability
            correct += pred.eq(target.view_as(pred)).sum().item()
    test_loss /= len(test_loader.dataset)

#     print('\nTest set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
#         test_loss, correct, len(test_loader.dataset),
#         100. * correct / len(test_loader.dataset)))

    return correct / len(test_loader.dataset)
````

````python
compression_ranks = np.linspace(1, 1000, 5, dtype=int)
avg_time_per_rank = []

for compression_rank in tqdm(compression_ranks):
    time = %timeit -oq test_accuracy(CompressedNet(model, compression_rank, layer_svd), test_loader)
    avg_time_per_rank.append(time.average)
````

Number of NN parameters depends on compression rank $r$. The lower compression rank leads to lower number of parameters and then to speed-up.

````python
import jax.numpy as jnp
import jax
import numpy as np
import random
from jax import jacfwd

class Game:
    def __init__(self, key):
        # key is a jax.random.PRNGKey
        self.key = key
        return
    def hermione_chooses_r_and_W(self):
        self.r = random.randint(1, 95)
        self.key, subkey = jax.random.split(self.key)
        self.W1 = jax.random.uniform(subkey, (self.r, 100), maxval=100., minval=0., dtype=jnp.float32)
        self.key, subkey = jax.random.split(self.key)
        self.W2 = jax.random.uniform(subkey, (100, self.r), maxval=100., minval=0., dtype=jnp.float32)
    def hermione_computes_function(self, x):
        return jnp.sin(self.W2 @ jnp.cos(self.W1 @ x))
    def harry_answers(self, m):
        # <your code here>
        # you shouldn't use self.r, self.W1, or self.W2
        # you can call `hermione_computes_function` multiple times
        x_sample = np.array([np.random.uniform(size=100, low=0, high=100) for _ in range(m)])
        func_grad_x = jax.jit(jacfwd(self.hermione_computes_function))
        grad_matrix = func_grad_x(x_sample[0])[0].reshape(100, 1)
        
        for i in range(1, m):
            grad_matrix = np.hstack((grad_matrix, func_grad_x(x_sample[i])[0].reshape(100,1)))
        
        grad_matrix = grad_matrix / jnp.sqrt(m)
        S = jnp.linalg.svd(grad_matrix)[1]
        
        return S[S > 1.].shape[0]
    
    def play(self, n_rounds, verbose=True):
        # n_rounds: a number of rounds of the game
        # verbose: print or not the result of each round
        n_right_answers = 0
        for _ in range(n_rounds):
            self.hermione_chooses_r_and_W()
            r = self.harry_answers(m=100)
            if abs(r - self.r) == 0:
                if verbose:
                    print("Good job! The true answer is {}, Harry's answer is {}!".format(self.r, r))
                n_right_answers += 1
            else:
                if verbose:
                    print("Harry's answers is {}, but the true answer is {} :(".format(r, self.r))
        if float(n_right_answers) / n_rounds > 0.95:
            print('Well done: {}/{} right answers!'.format(n_right_answers, n_rounds))
        else:
            print('Only {}/{} right answers :( Work a little more and you will succeed!'.format(
                n_right_answers, n_rounds))
````
